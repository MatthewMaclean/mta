```r
library('dplyr')
library('igraph')
```

```r
g = read_graph('../../data/processed/mta-elevators.graphml',format = 'graphml')
```

```r
V(g)$label[V(g)$node_type == 'Platform'] <- paste0("Platform pid_",c(1:sum(V(g)$node_type == 'Platform'))) 
```

```r
V(g)$name <- paste0(V(g)$station,":",V(g)$label)
el <- as_data_frame(g)


# write.csv(el,file='../../data/processed/edgelist_w_pid.csv',row.names=FALSE)
```

```r
from <- rep(NA,dim(el)[1])
to <- rep(NA,dim(el)[1])
station <- rep(NA,dim(el)[1])
from_type <- rep(NA,dim(el)[1])
to_type <- rep(NA,dim(el)[1])

for(i in c(1:dim(el)[1])){
    from[i] <- strsplit(el[i,1],":")[[1]][2]
    from_type[i] <- unique(V(g)$node_type[V(g)$label == from[i]])[1]
    to[i] <- strsplit(el[i,2],":")[[1]][2]
    to_type[i] <- unique(V(g)$node_type[V(g)$label == to[i]])[1]
    station[i] <- strsplit(el[i,1],":")[[1]][1]
}

el_w_st <- data.frame(station_name=station,from=from,from_type=from_type,to=to,to_type=to_type,stringsAsFactors = FALSE)
```

```r
head(el_w_st)
```

```r
write.csv(el_w_st,file='../../data/processed/edgelist_w_pid.csv',row.names=FALSE)
```
